```python
import cv2
from ultralytics import YOLO
from matplotlib import pyplot as plt
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model.predict(source="people.jpg", classes=0, conf=0.8, verbose=False)
```

```python
results[0].names
```

```python
results[0].boxes.shape[0]
```

```python
plot_res = results[0].plot(labels=True)
```

```python
plot_res = cv2.cvtColor(plot_res, cv2.COLOR_BGR2RGB)
plt.imshow(plot_res);
```
